```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

df = pd.read_excel('WDIW.xlsx')
```

```python
print(df.columns)
```

```python
print(df['Birth rate, crude (per 1,000 people)'].isnull().sum())
```

```python
df_dropped = pd.DataFrame.copy(df)

drop_rows = df_dropped[df_dropped['Hult Region'] != 'Eastern Europe']
drop_rows = drop_rows[drop_rows['Hult Region'] != 'Central Asia']
drop_rows = drop_rows[drop_rows['Hult Region'] != 'Australia & Oceania']
drop_rows = drop_rows[drop_rows['Hult Region'] != 'Carribean Islands & Parts of Latin America'] 
drop_rows = drop_rows[drop_rows['Hult Region'] != 'East/South Africa'] 
drop_rows = drop_rows[drop_rows['Hult Region'] != 'Carribean Islands & Parts of Latin America '] 
drop_rows = drop_rows[drop_rows['Hult Region'] != 'Northern Sub-Saharan Africa'] 
drop_rows = drop_rows[drop_rows['Hult Region'] != 'Middle East & Russia'] 
drop_rows = drop_rows[drop_rows['Hult Region'] != 'North America & Part of Latin America'] 
drop_rows = drop_rows[drop_rows['Hult Region'] != 'East/Southeast Asia'] 

```

```python
birth = drop_rows[['Hult Region', 'Birth rate, crude (per 1,000 people)',]]
print(birth['Birth rate, crude (per 1,000 people)'].isnull().sum())
birth = birth.dropna()

```

```python

sns.violinplot(x = 'Hult Region',
               y = 'Birth rate, crude (per 1,000 people)',
               data = birth,
               orient = 'v',
               inner = None,
               color = 'white')
sns.stripplot(x = 'Hult Region',
              y = 'Birth rate, crude (per 1,000 people)',
              data = birth,
              jitter = True,
              size = 5,
              orient = 'v')
plt.xlabel('')
plt.xticks(rotation=15)
plt.title('Birth rate (crude)')
sns.despine()
plt.ylabel('per 1,000 people')

plt.savefig("birth1.png")
plt.show()
```
